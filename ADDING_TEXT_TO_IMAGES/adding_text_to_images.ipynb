```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Create a empty canvas
height,width = 500,700
blue = (255,127,0)
canvas = np.full((height,width,3),blue,dtype=np.uint8)
canvas = cv2.cvtColor(canvas,cv2.COLOR_BGR2RGB)
plt.imshow(canvas)
```

```python
text = "Hello, OpenCV"
font = cv2.FONT_HERSHEY_COMPLEX
# org is the oordinate of the bottom left corner of the text
org = (50,100)
font_scale = 1.5
color = (255,255,255)
thickness = 4

text_canvas = cv2.putText(canvas,text,org,font,font_scale,color,thickness,cv2.LINE_AA)
plt.imshow(canvas)
```

```python
canvas = np.full((height,width,3),blue,dtype=np.uint8)
canvas = cv2.cvtColor(canvas,cv2.COLOR_BGR2RGB)

fonts = [
  cv2.FONT_HERSHEY_SIMPLEX,
  cv2.FONT_HERSHEY_COMPLEX,
  cv2.FONT_HERSHEY_SCRIPT_SIMPLEX,
  cv2.FONT_HERSHEY_SCRIPT_COMPLEX,
  cv2.FONT_HERSHEY_COMPLEX_SMALL
]

y_offset = 50
for i,font in enumerate(fonts):
  text = f"Font {i+1}"
  cv2.putText(canvas, text, (50,y_offset),font,1,(0,0,0),2,cv2.LINE_AA)
  y_offset+=50


plt.imshow(canvas)
```
