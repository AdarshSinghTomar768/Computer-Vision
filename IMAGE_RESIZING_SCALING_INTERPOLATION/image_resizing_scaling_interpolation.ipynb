```python
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("./bgr_image.png")
```

```python
rgb_image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(rgb_image)
```

### IMAGE RESIZING

```python
print(rgb_image.shape)
```

```python
new_height,new_width = 400,400
resized_image = cv2.resize(rgb_image,(new_width,new_height))
```

```python
plt.imshow(resized_image)
```

### IMAGE SCALING

```python
scale_percentage = 50
s_width = int(rgb_image.shape[1] * scale_percentage / 100)
s_height = int(rgb_image.shape[0] * scale_percentage / 100)
```

```python
scale_image = cv2.resize(rgb_image,(s_width,s_height),interpolation=cv2.INTER_AREA)
```

```python
plt.imshow(scale_image)
```

```python
print(scale_image.shape)
```

### INTERPOLATION

```python
resize_width = 600
resize_height = 600

# Interpolation methord
methods = {
  'Nearest' : cv2.INTER_NEAREST,
  'Linear' : cv2.INTER_LINEAR,
  'Cubic' : cv2.INTER_CUBIC,
  'Area' : cv2.INTER_AREA,
  'Lanczos4' : cv2.INTER_LANCZOS4
}
```

```python
resized_images = {}
for name, method in methods.items():
    resized = cv2.resize(rgb_image, (resize_width, resize_height), interpolation=method)
    resized_images[name] = resized
    cv2.imwrite(f"./{name}.jpg", resized)

plt.figure(figsize=(12, 8))

# Original Image
plt.subplot(2, 3, 1)
plt.imshow(rgb_image)
plt.title("Original Image")
plt.axis("off")

# Resized Images
for i, (name, resized) in enumerate(resized_images.items(), start=2):
    plt.subplot(2, 3, i)
    plt.imshow(resized)
    plt.title(name)
    plt.axis("off")

plt.tight_layout()
plt.show()
```
