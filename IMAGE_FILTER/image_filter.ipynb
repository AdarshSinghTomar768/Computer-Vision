```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("./filter.jpg",cv2.IMREAD_GRAYSCALE)

plt.imshow(image,cmap="grey")
```

```python
horizontal_kernel = np.array([[-1,-1,-1],
                       [ 0, 0, 0],
                       [ 1, 1, 1]])
```

```python
horizontal_filter_image = cv2.filter2D(image,-1,horizontal_kernel)
plt.imshow(horizontal_filter_image,cmap="grey")
```

```python
# More intensity of horizontal lines
x = -15
horizontal_kernel = np.array([[-1,-1,-1],
                       [ x, x, x],
                       [ 1, 1, 1]])

horizontal_filter_image = cv2.filter2D(image,-1,horizontal_kernel)
plt.imshow(horizontal_filter_image,cmap="grey")
```

```python
# Vertical lines
x = -15
vertical_kernel = np.array([[-1, x, 1],
                            [ -1, x, 1],
                            [-1, x, 1]])

vertical_filter_image = cv2.filter2D(image,-1,vertical_kernel)
plt.imshow(vertical_filter_image,cmap="grey")
```

```python
# Vertical lines
x = -15
diagional_kernel = np.array([[x,-1,-1],
                             [1, x,-1],
                             [1, 1, x]])

diagional_filter_image = cv2.filter2D(image,-1,diagional_kernel)
plt.imshow(diagional_filter_image,cmap="grey")
```

```python
plt.Figure(figsize=(10,10))
plt.subplot(2,2,1)
plt.title("Original Image")
plt.imshow(image,cmap="grey")

plt.subplot(2,2,2)
plt.title("Horizontal Filter Image")
plt.imshow(horizontal_filter_image,cmap="grey")

plt.subplot(2,2,3)
plt.title("Vertical Filter Image")
plt.imshow(vertical_filter_image,cmap="grey")

plt.subplot(2,2,4)
plt.title("Diagional Filter Image")
plt.imshow(diagional_filter_image,cmap="grey")

plt.show()
```
