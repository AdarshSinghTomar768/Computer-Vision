```python
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("./bgr_image.png")
```

```python
plt.imshow(image)
```

```python
image_rgb = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
```

# FLIP

```python
# 1 --> Horizontally
# 0 --> Vercically
# -1 --> Both Horizontally and Vertically

flip_horizontal = cv2.flip(image,1)
flip_vertically = cv2.flip(image,0)
flip_both = cv2.flip(image,-1)

titles = ["Original image","Horizontal_flip","Vertical_flip","Flip_Both_Axis"]
images = [image_rgb, 
          cv2.cvtColor(flip_horizontal,cv2.COLOR_BGR2RGB), 
          cv2.cvtColor(flip_vertically,cv2.COLOR_BGR2RGB), 
          cv2.cvtColor(flip_both,cv2.COLOR_BGR2RGB)]


plt.Figure(figsize=(12,8))
for i in range(4):
  plt.subplot(2,2,i+1)
  plt.imshow(images[i])
  plt.title(titles[i])
  plt.axis("off")

plt.tight_layout()
plt.show()
```

# ROTATE

```python
rotate_90_clockwise = cv2.rotate(image_rgb,cv2.ROTATE_90_CLOCKWISE)
rotate_90_counterclockwise = cv2.rotate(image_rgb,cv2.ROTATE_90_COUNTERCLOCKWISE)
rotate_180 = cv2.rotate(image_rgb,cv2.ROTATE_180)

titles = ["Original image","90_degree_clockwise","90_degree_counterclockwise","180_degree"]
images = [image_rgb, 
          rotate_90_clockwise, 
          rotate_90_counterclockwise, 
          rotate_180]

plt.Figure(figsize=(12,8))
for i in range(4):
  plt.subplot(2,2,i+1)
  plt.imshow(images[i])
  plt.title(titles[i])
  plt.axis("off")

plt.tight_layout()
plt.show()
```

# CROPPING AND IMAGE

```python
print(image_rgb.shape)
print(type(image_rgb))
```

```python
x_start, y_start = 50,100
x_end, y_end = 400,250
```

```python
cropped_image = image_rgb[y_start:y_end , x_start:x_end]

plt.Figure(figsize=(8,4))

plt.subplot(2,2,1)
plt.imshow(image_rgb)
plt.title("Original_image")
plt.axis("off")

plt.subplot(2,2,2)
plt.imshow(cropped_image)
plt.title("Cropped_image")
plt.axis("off")

plt.tight_layout()
plt.show()
```
