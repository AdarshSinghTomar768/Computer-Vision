```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Create a empty canvas
height,width = 500,700
blue = (255,127,0)
canvas = np.full((height,width,3),blue,dtype=np.uint8)

plt.imshow(cv2.cvtColor(canvas,cv2.COLOR_BGR2RGB))
```

# DRAW A LINE

```python
line = cv2.line(canvas,(50,50),(650,450),(255,255,255),5)
```

```python
plt.imshow(cv2.cvtColor(line,cv2.COLOR_BGR2RGB))
```

# DRAW A RECTANGLE

```python
rectangle = cv2.rectangle(canvas,(100,100),(300,300),(0,255,255),5)
plt.imshow(cv2.cvtColor(rectangle,cv2.COLOR_BGR2RGB))
```

```python
rectangle = cv2.rectangle(canvas,(100,100),(300,300),(0,255,255),-1)
plt.imshow(cv2.cvtColor(rectangle,cv2.COLOR_BGR2RGB))
```

# DRAW A CIRCLE

```python
circle = cv2.circle(canvas,(500,150),100,(0,0,255),5)
plt.imshow(cv2.cvtColor(circle,cv2.COLOR_BGR2RGB))
```

```python
circle = cv2.circle(canvas,(500,150),100,(0,0,255),-1)
plt.imshow(cv2.cvtColor(circle,cv2.COLOR_BGR2RGB))
```

# DRAW AN ELLIPSE

```python
ellispe = cv2.ellipse(canvas,(350,350),(100,50),45,0,360,(0,255,0),5)
plt.imshow(cv2.cvtColor(ellispe,cv2.COLOR_BGR2RGB))
```

# DRAW A POLYGON

```python
# Xn, Yn where n>2
points = np.array([[150,400],[300,400],[225,300]],np.int32)
new_reshaped_points = points.reshape((-1,1,2))

print(new_reshaped_points)
polygon = cv2.polylines(canvas,[new_reshaped_points],True,(255,0,255),3)

plt.imshow(cv2.cvtColor(polygon,cv2.COLOR_BGR2RGB))
```
