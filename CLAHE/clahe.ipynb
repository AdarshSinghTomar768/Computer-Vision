```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = cv2.imread("./bgr_image.png")
rgb_image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(rgb_image)
```

```python
grey_image = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
plt.imshow(grey_image,cmap='grey')
```

```python
equalized_image = cv2.equalizeHist(grey_image)
clahe = cv2.createCLAHE(clipLimit=2.0,tileGridSize=(4,4))
clahe_equalized_image = clahe.apply(grey_image)

plt.figure(figsize=(12,8))


plt.subplot(1,3,1)
plt.imshow(grey_image,cmap='grey')
plt.title("Original Image")

plt.subplot(1,3,2)
plt.imshow(equalized_image,cmap='grey')
plt.title("Equalised Image")

plt.subplot(1,3,3)
plt.imshow(clahe_equalized_image,cmap='grey')
plt.title("CLAHE Equalised Image")

plt.tight_layout()
plt.show()
```

```python
yuv_image = cv2.cvtColor(image,cv2.COLOR_BGR2YUV)
yuv_image[:,:,0] = clahe.apply(yuv_image[:,:,0])
equalized_color_image = cv2.cvtColor(yuv_image,cv2.COLOR_YUV2BGR)
equalized_color_image_rgb = cv2.cvtColor(equalized_color_image,cv2.COLOR_BGR2RGB)

plt.figure(figsize=(12,8))


plt.subplot(1,2,1)
plt.imshow(rgb_image)
plt.title("Original Image")

plt.subplot(1,2,2)
plt.imshow(equalized_color_image_rgb)
plt.title("CLAHE Equalised RGB Image")

plt.tight_layout()
plt.show()
```
