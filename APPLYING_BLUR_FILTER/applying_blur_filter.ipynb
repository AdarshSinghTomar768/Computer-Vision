```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = cv2.imread("./bgr_image.png")
rgb_image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
grey_image = rgb_image = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(rgb_image)
```

```python
kernel_s = 5
average_blur = cv2.blur(rgb_image,(kernel_s,kernel_s))
plt.imshow(average_blur)
```

```python
kernel_s = 5
average_blur = cv2.blur(grey_image,(kernel_s,kernel_s))
plt.imshow(average_blur,cmap='grey')
```

```python
kernel_s = 7
guassian_blur = cv2.GaussianBlur(grey_image,(kernel_s,kernel_s),0)
plt.imshow(guassian_blur,cmap='grey')
```

```python
kernel_s = 5
median_blur = cv2.medianBlur(grey_image,kernel_s)
plt.imshow(median_blur,cmap='grey')
```

```python
plt.figure(figsize=(14,12))

plt.subplot(2,2,1)
plt.title("Original_Grey_Image")
plt.imshow(grey_image,cmap='grey')
plt.axis('off')

plt.subplot(2,2,2)
plt.title("Average_blur_Image")
plt.imshow(average_blur,cmap='grey')
plt.axis('off')

plt.subplot(2,2,3)
plt.title("Guassian_blur_Image")
plt.imshow(guassian_blur,cmap='grey')
plt.axis('off')

plt.subplot(2,2,4)
plt.title("Median_blur_Image")
plt.imshow(median_blur,cmap='grey')
plt.axis('off')
```
