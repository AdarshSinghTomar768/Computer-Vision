```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = cv2.imread("./bgr_image.png")
rgb_image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(rgb_image)
```

```python
grey_image = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
plt.imshow(grey_image,cmap='grey')
```

```python
hist, bins = np.histogram(grey_image.flatten(),256,[0,256])
```

```python
equalized_image = cv2.equalizeHist(grey_image)

equalized_hist, bins = np.histogram(equalized_image.flatten(),256,[0,256])


plt.figure(figsize=(12,6))

plt.subplot(2,2,1)
plt.imshow(grey_image,cmap='grey')
plt.title("Original Image")
# plt.axis("off")

plt.subplot(2,2,2)
plt.plot(hist,color='grey')
plt.title("Original Image Histogram")
# plt.axis("off")

plt.subplot(2,2,3)
plt.imshow(equalized_image,cmap='grey')
plt.title("Equalized Image")
# plt.axis("off")

plt.subplot(2,2,4)
plt.plot(equalized_hist,color='green')
plt.title("Equalized Image Histogram")
# plt.axis("off")

plt.tight_layout()
plt.show()
```

```python
yuv_image = cv2.cvtColor(image,cv2.COLOR_BGR2YUV)

yuv_image[:,:,0] = cv2.equalizeHist(yuv_image[:,:,0])

equalized_color_image = cv2.cvtColor(yuv_image,cv2.COLOR_YUV2BGR)

equalized_color_image_rgb = cv2.cvtColor(equalized_color_image,cv2.COLOR_BGR2RGB)
```

```python
plt.figure(figsize=(12,8))

plt.subplot(1,2,1)
plt.imshow(rgb_image)
plt.title("Original Image")
# plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(equalized_color_image_rgb)
plt.title("Equalized RGB Image Histogram")
# plt.axis("off")


plt.tight_layout()
plt.show()
```
