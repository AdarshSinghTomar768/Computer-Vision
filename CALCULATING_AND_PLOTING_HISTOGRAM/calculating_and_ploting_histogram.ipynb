```python
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("./bgr_image.png")
rgb_image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(rgb_image)
```

```python
channels = cv2.split(rgb_image)
colors = ['red','green','blue']

for channel,color in zip(channels,colors):
  hist = cv2.calcHist([channel],[0],None,[256],[0,256])
  plt.plot(hist,color=color)

plt.title("Color Histogram")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")
plt.show()
```

```python
hsv_image = cv2.cvtColor(image,cv2.COLOR_BGR2HSV)

h, s, v = cv2.split(rgb_image)



h_hist = cv2.calcHist([h],[0],None,[180],[0,180])
s_hist = cv2.calcHist([s],[0],None,[256],[0,256])
v_hist = cv2.calcHist([v],[0],None,[256],[0,256])

fig, axis = plt.subplots(1,3, figsize=(15,5))

axis[0].plot(h_hist,color="orange")
axis[0].set_title("Hue Histogram")
axis[0].set_xlabel("Bins")
axis[0].set_ylabel("Frequency")

axis[1].plot(s_hist,color="green")
axis[1].set_title("Saturation Histogram")
axis[1].set_xlabel("Bins")
axis[1].set_ylabel("Frequency")

axis[2].plot(v_hist,color="blue")
axis[2].set_title("Value Histogram")
axis[2].set_xlabel("Bins")
axis[2].set_ylabel("Frequency")


plt.tight_layout()
plt.show()
```
