```python
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("./bgr_image.png")
rgb_image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
grey_image = rgb_image = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
```

```python
# Sobel edge detector

sobel_x = cv2.Sobel(grey_image,cv2.CV_64F,1,0,ksize=3)
sobel_y = cv2.Sobel(grey_image,cv2.CV_64F,1,0,ksize=3)

sobel_combined = cv2.magnitude(sobel_x,sobel_y)
```

```python
# Laplacian edge detector

laplacian = cv2.Laplacian(grey_image,cv2.CV_64F,ksize=3)
laplacian_abs = cv2.convertScaleAbs(laplacian)
```

```python
# Canny edge detection
canny_edge = cv2.Canny(grey_image,180,200)
```

```python
titles = ["Original","Sobel","Laplacian all","Laplacian","Canny"]
images = [grey_image,sobel_combined,laplacian,laplacian_abs,canny_edge]

plt.figure(figsize=(20,16))
for i in range(len(titles)):
  plt.subplot(3,2,i+1)
  plt.imshow(images[i],cmap="grey")
  plt.title(titles[i])
  plt.axis("off")

plt.show()
```
